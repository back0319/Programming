```python
def find_even(L):
  BL=[]
  print("원소의 개수: ", len(L))
  for i in range(len(L)):   # for (i=0; i<len(L); i++)의 의미를 가짐.
    if L[i] % 2 == 0:
      BL.append(True)
    else:
      BL.append(False)
  return BL

Data = np.array([12, 31, 24, 35, 67, 10])
B = find_even(Data)
print(B)
print(Data[B])
print("*"*40)
Data = pd.Series([22, -1, 12, 3, 6])
B = find_even(Data)
print(B)
print(Data[B])
```
