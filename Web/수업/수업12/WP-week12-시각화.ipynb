## 피봇 테이블과 히트맵

```python
import seaborn as sns
titanic=sns.load_dataset('titanic')
titanic_size = titanic.pivot_table(index='class', columns='sex', aggfunc='size')
titanic_size
```

```python
import pandas as pd
ts = pd.DataFrame(titanic.groupby(['class'])[['sex']].value_counts())
ts
```

```python
ts.unstack()
```
